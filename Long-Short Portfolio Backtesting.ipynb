```python
import pandas as pd
import numpy as np
import yfinance as yf
from scipy.optimize import minimize
import matplotlib.pyplot as plt

#Step 1: Choose an ETF with a minimum of 100 assets, identify those assets
#Load ETF holdings from a CSV file
holdings_df = pd.read_csv("D:\\Holdings_details_S&P_500_ETF.csv")
holdings_df.columns = holdings_df.columns.str.strip()
tickers = holdings_df['TICKER'].tolist()

tickers = [ticker for ticker in tickers if isinstance(ticker, str) and ticker.strip() != '']

# Step 2: Retrieve historical price data for the selected assets
def get_historical_data(tickers, start_date, end_date):
    data = yf.download(tickers, start=start_date, end=end_date)
    return data['Close']

# Get 5 years of historical data for the assets
historical_data = get_historical_data(tickers, start_date='2018-01-01', end_date='2023-12-31')
# Check if the tickers list is valid
valid_tickers = historical_data.columns.tolist()
print(f"Valid tickers list: {tickers}")

tickers = [ticker for ticker in tickers if ticker in valid_tickers]

# Calculate the daily returns and covariance matrix
returns = historical_data[tickers].pct_change().dropna() 
cov_matrix = returns.cov()  

# Define the portfolio volatility function
def portfolio_volatility(weights, cov_matrix):
    return np.sqrt(np.dot(weights.T, np.dot(cov_matrix, weights)))

# Initialize weights for the optimization (equal weight for each asset)
init_weights = np.ones(len(tickers)) / len(tickers)

# Set bounds and constraints for the optimization
bounds = tuple((0, 1) for _ in range(len(tickers)))  
constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1})

# https://docs.scipy.org/doc/scipy/reference/optimize.minimize-slsqp.html
# Optimize the portfolio weights using Sequential Least Squares Programming
result = minimize(portfolio_volatility, init_weights, args=(cov_matrix,), method='SLSQP', bounds=bounds, constraints=constraints)
optimal_weights = result.x
print("Optimal portfolio weights:", optimal_weights)

# Calculate the expected portfolio return and volatility
portfolio_return = np.sum(returns.mean() * optimal_weights) 
portfolio_vol = portfolio_volatility(optimal_weights, cov_matrix) 

print(f"Expected portfolio return: {portfolio_return:.4f}")
print(f"Portfolio volatility: {portfolio_vol:.4f}")

# Step 4-5
momentum = historical_data.pct_change(periods=126).iloc[-1] 
z_scores = (momentum - momentum.mean()) / momentum.std() 

# Select long and short baskets based on z-scores
long_basket = z_scores[z_scores > 1].index.tolist() 
short_basket = z_scores[z_scores < -1].index.tolist() 

# Calculate monthly returns for long, short, and ETF
long_returns = historical_data[long_basket].pct_change().mean(axis=1)
short_returns = historical_data[short_basket].pct_change().mean(axis=1)
etf_returns = historical_data.mean(axis=1).pct_change()
monthly_returns = (long_returns - short_returns) 

# Plot the monthly portfolio return and cumulative return vs ETF
plt.figure(figsize=(12, 6))

# Monthly Portfolio Return (with positive/negative coloring)
plt.subplot(3, 1, 1)
plt.bar(monthly_returns.index, monthly_returns, color=['green' if x > 0 else 'red' for x in monthly_returns])
plt.title('Monthly Portfolio Return (Long-Short Strategy)')
plt.ylabel('Monthly Return')

# Monthly return for Long Picks vs Short Picks vs ETF
plt.subplot(3, 1, 2)
plt.plot(long_returns.index, long_returns, label='Long Picks', color='green')
plt.plot(short_returns.index, short_returns, label='Short Picks', color='red')
plt.plot(etf_returns.index, etf_returns, label='ETF', color='blue')
plt.title('Monthly Return: Long Picks vs Short Picks vs ETF')
plt.legend()

# Cumulative Portfolio Return vs ETF
cumulative_returns = (1 + monthly_returns).cumprod() - 1
cumulative_etf = (1 + etf_returns).cumprod() - 1
plt.subplot(3, 1, 3)
plt.plot(cumulative_returns.index, cumulative_returns, label='Portfolio Cumulative Return', color='purple')
plt.plot(cumulative_etf.index, cumulative_etf, label='ETF Cumulative Return', color='blue')
plt.title('Cumulative Portfolio Return vs ETF')
plt.legend()

plt.tight_layout()
plt.show()
```
